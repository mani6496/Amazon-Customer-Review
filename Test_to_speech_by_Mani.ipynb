```python
from gtts import gTTS
import IPython
```

```python
def text_to_speech(text):
    speech = gTTS(text = text, lang = "en", slow = False)
    speech.save("output.mp3")
    return IPython.display.Audio("output.mp3", autoplay=True)
```

```python
text = "I am a chatgpt bot and understand any human's language.I quick response any question and show answer with related question."
```

```python
text_to_speech(text)
```
